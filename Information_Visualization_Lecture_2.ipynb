```python
import random

population = list(range(1,101))
sample_size= 10
random_sample = random.sample(population, sample_size)
print("Random Sample:",random_sample)
```

```python
import numpy as np
import pandas as pd

data = pd.DataFrame({
    "Category" : [ 'A', 'B', 'C', 'D'] * 25,
    'Value': np.random.randn(100)
                 })
stratified_sample = data.groupby('Category', group_keys=False).apply(lambda x: x.sample(2))
print("Stratified Sample: \n", stratified_sample)
```

```python
clusters = [list(range(1,26)), list(range(26,51)), list(range(51,76)), list(range(76,101))]
cluster_sample = random.choice(clusters)

print("Cluster Sample:",cluster_sample)
```

```python
###Descriptive Statistics
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Values': np.random.randn(100)
})

summary = data.describe()
print("Summary Statistics:\n", summary)
```

```python
###Data Visualization

import matplotlib.pyplot as plt

data['Values'].hist(bins=10, color='red', edgecolor='black')
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
### Data Samplingand Dimensionality Reduction
### Principal Component Analysis (PCA)

from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(iris.data)

plt.scatter(reduced_data[:,0], reduced_data[:,1], c=iris.target, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label='Species')
plt.show()
```

```python
#parallel coordinates
from pandas.plotting import parallel_coordinates
data=pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv')
parallel_coordinates(data,'class',color=['b','g','r'])
plt.title('Parallel coordinates Plot')
plt.xlabel('Attributes')
plt.ylabel("Values")
plt.show()
```

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({
    'Year': [2000, 2001, 2002, 2003, 2004, 2005],
    'Value': [100, 200, 150, 300, 250, 400],
    'Country': ['USA']*6
})

fig = px.line(data, x='Year', y='Value', title='Time Series Data Visualization', labels={'Value':'Sample Value'})
fig.show()
```
